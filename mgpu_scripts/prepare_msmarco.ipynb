```python
import pandas as pd

all_df=pd.read_csv('./msmarco/msmarco_MAB_100docs.tsv', delimiter="\t") 
```

```python
all_df
```

```python
# build query dictionary
query_dict={}
doc_dict={}

query_dict_key={}
doc_dict_key={}

for index, rows in all_df.iterrows():
    if rows['qid'] not in query_dict:
        query_dict[rows['qid']]=rows['query']
    if rows['docid'] not in doc_dict:
        doc_dict[rows['docid']]=rows['article']
```

```python
import numpy as np
association_mat=np.zeros((len(query_dict),len(doc_dict)))

query_count=-1
doc_count=-1

for index, rows in all_df.iterrows():
    if rows['qid'] not in query_dict_key:
        query_count+=1
        query_dict_key[rows['qid']]=query_count
    if rows['docid'] not in doc_dict_key:
        doc_count+=1
        doc_dict_key[rows['docid']]=doc_count
```

```python
association_mat
```

```python
np.sum(association_mat, axis=1)
```

```python
for index, rows in all_df.iterrows():
    #print(rows['qid'], rows['docid'])
    association_mat[query_dict_key[rows['qid']],doc_dict_key[rows['docid']]]=1
    
```

```python
corpus_query=[]
corpus_doc=[]

for key, value in query_dict.items():
    corpus_query.append(value)
    
for key, value in doc_dict.items():
    corpus_doc.append(value)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer,CountVectorizer

#instantiate CountVectorizer()
cv=CountVectorizer(min_df=0.1, max_df=0.7, max_features=500, analyzer= 'word', stop_words='english') 
tfidfvectorizer = TfidfVectorizer(max_features=500, analyzer='word',stop_words= 'english')

question_tfidf_vector=tfidfvectorizer.fit_transform(corpus_query)
page_tfidf_text_vector=tfidfvectorizer.fit_transform(corpus_doc)


# # this steps generates word counts for the words in your docs 
# page_count_text_vector=cv.fit_transform(corpus_text)


# page_count_title_vector=cv.fit_transform(corpus_title)
# page_tfidf_title_vector=tfidfvectorizer.fit_transform(corpus_title)

# page_count_keywords_vector=cv.fit_transform(corpus_keywords)
# page_tfidf_keywords_vector=tfidfvectorizer.fit_transform(corpus_keywords)

# page_count_description_vector=cv.fit_transform(corpus_description)
# page_tfidf_description_vector=tfidfvectorizer.fit_transform(corpus_description)

# question_count_vector=cv.fit_transform(corpus_question)
```

```python
from sklearn.linear_model import Ridge
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics.pairwise import linear_kernel
from sklearn.kernel_approximation import Nystroem
from sklearn.model_selection import KFold
import pickle
import random
import numpy as np
from itertools import product
from sklearn import metrics

ndim=768

query_f = pickle.load(open('./msmarco/X_question_bert.pkl','rb'))
page_f = pickle.load(open('./msmarco/X_page_bert.pkl','rb'))
association_mat = pickle.load(open('./msmarco/X_ass_mat.pkl','rb'))


    
def prepare_context(query_id, page_id):
    #generate the context, the dimension is number of features * number of arms
    context =np.zeros(ndim*2)

    context[0:ndim] = query_f[query_id,:]
    context[ndim:2*ndim] = page_f[page_id,:]

    #print(len(context))
    return context



total_sample_n = 34100

Y = np.zeros(total_sample_n)
X = np.zeros((total_sample_n, ndim*2))

label_true = np.zeros((341,100))
predict_score = np.zeros((341,100))

count=0

kf = KFold(n_splits=5)


for i in range(341):
    for j in range(100):
        X[count,:]=prepare_context(i,j)
        Y[count]=association_mat[i,j]
        count+=1
```

```python
n_c=500
feature_map_nystroem = Nystroem(kernel='rbf', gamma=0.05, random_state=1, n_components=n_c)
data_transformed = feature_map_nystroem.fit_transform(X)
pickle.dump(data_transformed, open('./msmarco/K_arm_rbf_005_bert.pkl','wb'))
```
